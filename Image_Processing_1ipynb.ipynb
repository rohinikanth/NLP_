Resize img using Pillow

```python
from PIL import Image
img = Image.open('J.jpg')
img_res = img.resize((200,160))
print(img_res.size)
print(img_res.mode)
print(img_res.format)
print(type(img_res))
img_res.save('J_rez.jpg')
plt.imshow(img_res)
plt.show()
```

RGB to Grayscale conversion using opencv

```python
import cv2
img = cv2.imread('J_rez.jpg')
print(type(img))
print(img.shape)
print(img)
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
from google.colab.patches import cv2_imshow
cv2_imshow(grayscale_img)
```
